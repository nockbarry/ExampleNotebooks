```python
from sklearn.metrics import fbeta_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from scipy.sparse import hstack
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

def count_vowels(string):
    """Count the number of vowels in a string."""
    return sum(1 for char in string if char in 'aeiouAEIOU')

def count_consonants(string):
    """Count the number of consonants in a string."""
    return sum(1 for char in string if char.isalpha() and char not in 'aeiouAEIOU')

def count_digits(string):
    """Count the number of digits in a string."""
    return sum(1 for char in string if char.isdigit())

def count_special_characters(string):
    """Count the number of special characters in a string."""
    return len(string) - count_vowels(string) - count_consonants(string) - count_digits(string)

def compute_entropy(string):
    """Compute the Shannon entropy of a string."""
    # Get the probability of each character in the string
    prob = [string.count(c) / len(string) for c in set(string)]
    # Compute the entropy
    entropy = -sum([p * np.log2(p) for p in prob])
    return entropy

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score


# Compute entropy for each email name
email_data['entropy'] = email_data['email_name'].apply(compute_entropy)


# Display the first few rows with new features
email_data.head()


# 1. Feature Extraction

# Extract TF-IDF features
tfidf_vectorizer = TfidfVectorizer(analyzer='char', ngram_range=(2, 3))
X_tfidf = tfidf_vectorizer.fit_transform(email_data['email_name'])

# Extract additional features
email_data['num_vowels'] = email_data['email_name'].apply(count_vowels)
email_data['num_consonants'] = email_data['email_name'].apply(count_consonants)
email_data['length'] = email_data['email_name'].apply(len)
email_data['num_digits'] = email_data['email_name'].apply(count_digits)
email_data['num_special_chars'] = email_data['email_name'].apply(count_special_characters)

X_features = email_data[['num_vowels', 'num_consonants', 'length', 'num_digits', 'num_special_chars', 'entropy']]
X_combined = hstack([X_tfidf, X_features])

# 2. Data Splitting and Model Training

y = email_data['label']
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_combined, y, test_size=0.2, random_state=42)
# Standardize the features
scaler = StandardScaler(with_mean=False)  # with_mean=False is required for sparse matrices
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize classifiers
classifiers = {
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Machine": SVC(probability=True),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(50, 30), max_iter=500)
}

results = {}
mislabeled_data = {}

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    
    # Accuracy
    train_acc = clf.score(X_train, y_train)
    test_acc = clf.score(X_test, y_test)
    
    # F1 and F2 scores
    f1_score = fbeta_score(y_test, y_pred_test, beta=1)
    f2_score = fbeta_score(y_test, y_pred_test, beta=2)
    
    results[name] = {"Training Accuracy": train_acc, "Test Accuracy": test_acc, "F1 Score": f1_score, "F2 Score": f2_score}
    
    # Mislabeled instances for each model
    mislabeled_indices = y_test[y_test != y_pred_test].index
    mislabeled_data[name] = email_data.loc[mislabeled_indices][['email_name', 'label']]
    mislabeled_data[name]['predicted'] = y_pred_test[y_test != y_pred_test]

results_df = pd.DataFrame(results).T
results_df, mislabeled_data
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.neighbors import LocalOutlierFactor
from sklearn.covariance import EllipticEnvelope
from sklearn.metrics import accuracy_score, f1_score, fbeta_score

# Standardize the data
X_scaled = scaler.fit_transform(X_combined)

# As these are unsupervised methods, we'll treat all data as "unlabeled" and then evaluate against the actual labels
anomaly_detectors = {
    "Isolation Forest": IsolationForest(contamination=0.1),
    "One-Class SVM": OneClassSVM(nu=0.1, kernel="rbf"),
    "Local Outlier Factor": LocalOutlierFactor(contamination=0.1),
    "Elliptic Envelope": EllipticEnvelope(contamination=0.1)
}

results_anomaly = {}

for name, clf in tqdm(anomaly_detectors.items()):
    if name == "Local Outlier Factor":  # LOF doesn't have a 'predict' method
        y_pred = clf.fit_predict(X_scaled)
    else:
        clf.fit(X_scaled.toarray())
        y_pred = clf.predict(X_scaled)
    
    # Convert anomalies from -1 to 1 and normal from 1 to 0 to match our label encoding
    y_pred = [1 if x == -1 else 0 for x in y_pred]

    acc = accuracy_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    f2 = fbeta_score(y, y_pred, beta=2)

    results_anomaly[name] = {
        "Accuracy": acc,
        "F1 Score": f1,
        "F2 Score": f2
    }

results_anomaly_df = pd.DataFrame(results_anomaly).T
print(results_anomaly_df)
```

```python
from sklearn.cluster import DBSCAN, KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
from keras.layers import Input, Dense
from keras.models import Model
#from pyod.models.hbos import HBOS

# ... [Data Loading and Pre-processing] ...

# DBSCAN
db = DBSCAN(eps=0.5, min_samples=5).fit(X_scaled)
labels = db.labels_

# Autoencoders
input_layer = Input(shape=(X_scaled.shape[1],))
encoded = Dense(100, activation='relu')(input_layer)
encoded = Dense(50, activation='relu')(encoded)
decoded = Dense(50, activation='relu')(encoded)
decoded = Dense(100, activation='relu')(decoded)
output_layer = Dense(X_scaled.shape[1], activation='relu')(decoded)

autoencoder = Model(input_layer, output_layer)
autoencoder.compile(optimizer="adadelta", loss="mse")
autoencoder.fit(X_scaled, X_scaled, epochs=50, batch_size=256, shuffle=True, validation_split=0.2)
predictions = autoencoder.predict(X_scaled)
mse = np.mean(np.power(X_scaled - predictions, 2), axis=1)

# # HBOS
# hbos = HBOS()
# hbos.fit(X_scaled)
# hbos_scores = hbos.decision_scores_

# KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_scaled)
distances = kmeans.transform(X_scaled)
kmeans_scores = distances.min(axis=1)

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)
reconstructed = pca.inverse_transform(pca_result)
mse_pca = np.mean(np.power(X_scaled - reconstructed, 2), axis=1)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, precision_recall_curve, auc

# Define models and their scores
models = {
    "DBSCAN": labels,
    "Autoencoders": mse,
    "HBOS": hbos_scores,
    "KMeans": kmeans_scores,
    "PCA": mse_pca
}

# Assuming `y` is your ground truth
results = {}

for name, scores in models.items():
    # Determine threshold (here using top 5% scores)
    threshold = np.percentile(scores, 95)

    # Convert scores above threshold to anomalies (1)
    predictions = [1 if score > threshold else 0 for score in scores]

    # Calculate metrics
    acc = accuracy_score(y, predictions)
    precision = precision_score(y, predictions)
    recall = recall_score(y, predictions)
    f1 = f1_score(y, predictions)
    f2 = fbeta_score(y, predictions, beta=2)
    roc_auc = roc_auc_score(y, scores)
    pr_curve = precision_recall_curve(y, scores)
    pr_auc = auc(pr_curve[1], pr_curve[0])

    results[name] = {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "F2 Score": f2,
        "ROC AUC": roc_auc,
        "PR AUC": pr_auc
    }

results_df = pd.DataFrame(results).T
print(results_df)
```

# Active Learning

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Split the data into an initial training set and an unlabeled set
train_data, unlabeled_data = train_test_split(email_data, test_size=0.7, stratify=email_data['label'], random_state=42)

train_data.shape, unlabeled_data.shape

# Number of iterations for the active learning loop
iterations = 5

# Number of samples to query in each iteration
query_batch_size = 1000

# Lists to keep track of model performance over iterations
accuracies = []

# Active learning loop
for i in range(iterations):
    # Step 1: Train the One-Class SVM on the labeled data
    X_train = vectorizer.transform(train_data['email_name'])
    one_class_svm = OneClassSVM(kernel="linear", nu=0.1)
    one_class_svm.fit(X_train[train_data['label'] == 0])  # Train only on non-anomalous data (label=0)
    
    # Step 2: Predict on the unlabeled data
    X_unlabeled = vectorizer.transform(unlabeled_data['email_name'])
    decision_function_values = one_class_svm.decision_function(X_unlabeled)
    
    # Step 3: Select instances the model is least confident about
    uncertain_indices = np.argsort(np.abs(decision_function_values))[:query_batch_size]
    
    # Step 4: Query oracle for labels (simulated using true labels)
    queried_samples = unlabeled_data.iloc[uncertain_indices]
    
    # Step 5: Add queried samples to the training data and remove them from the unlabeled data
    train_data = pd.concat([train_data, queried_samples])
    unlabeled_data = unlabeled_data.drop(index=queried_samples.index)
    
    # Step 6: Evaluate the model performance on the entire dataset
    X_all = vectorizer.transform(email_data['email_name'])
    all_preds = one_class_svm.predict(X_all)
    all_preds = [1 if pred == -1 else 0 for pred in all_preds]
    accuracy = np.mean(all_preds == email_data['label'])
    accuracies.append(accuracy)

accuracies
```
